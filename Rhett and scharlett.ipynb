```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the non-dimensionalized equations
def dxdt(R, S, A, k):
    return -R + A + k * S * np.exp(-S)
    
def dydt(R, S, A, k):
    return -S + A + k * R * np.exp(-R)
```

```python
# Define the parameter values

A_S = 1.2
R_S = 1.0
k= 15
```

```python
# Define the grid of initial conditions
R0, S0 = np.meshgrid(np.linspace(0, 10, 30), np.linspace(0, 10, 30))
```

```python
# Compute the phase plane trajectories
dx = dxdt(R0, S0, A_S, k)
dy = dydt(R0, S0, R_S, k)
```

```python
# Plot the phase plane diagram
plt.figure(figsize=(8, 8))
plt.quiver(R0, S0, dx, dy, color='blue', alpha=0.5)
plt.xlabel('R')
plt.ylabel('S')
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.grid()
plt.show()
```
